#### 6. Plot both `y=x^2` and `y=exp(x)` in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

Syntax:

    Axes.set_xscale(value, **kwargs)
        Parameters: value : {"linear", "log", "symlog", "logit", ...}
    
    Axes.set_xlim(left=None, right=None, emit=True, auto=False, *, xmin=None, xmax=None)

````python
y = x^2
ye = exp(x)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize= (10,5))

ax1.set_xscale("linear") # Syntax doesn't say, but 'linear' seems to be the default, so this line is not needed.
ax1.set_xlim(85, 100)
ax1.set_yscale('linear')
ax1.set_xlabel('eje X')
ax1.grid()
ax1.plot(x, y, 'g')
ax1.plot(x, ye, 'r--')

ax2.set_xscale("linear")
ax2.set_yscale("log")
ax2.set_xlabel('eje X')
ax2.grid()
ax2.plot(x, y, 'g')
ax2.plot(x, ye, 'r--')
````
