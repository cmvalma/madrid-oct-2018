#### Whether the bin size affects your plot? Why?

Bins do affect the plot in the sense that bigger bings are too broad, too general, not reflecting substantial differences among contiguous ages, like for example the bigger bins show that people in their twenties and thirties are the most frequent, and thatś true, but there are many differences if we are interested in knowing about a particular age li 20, which looks to be among the most unfrequent age. There are many differences in ages. So, bin 50 looks too gross, but depending on our interest, we might want to have bins of 10 or 20...
