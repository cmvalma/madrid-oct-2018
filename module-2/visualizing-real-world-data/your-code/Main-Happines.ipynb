## Bin sizes  


<div style='color:darkblue'>
    <p></p>
    
There are some commonly-used rules-of-thumb that have been proposed (e.g. the Freedman-Diaconis Rule, Sturges' Rule, Scott's Rule, the Square-root rule, etc.) each of which has its own strengths and weaknesses.

A nice Python implementation of a variety of these auto-tuning histogram rules, in the histogram functionality in the latest version of the ***AstroPy package**... This works just like plt.hist, but lets you use syntax like, e.g. hist(x, bins='freedman') for choosing bins via the Freedman-Diaconis rule mentioned above.

From **matplotlib** version 2.0 or later and **numpy** version 1.11 or later, you can now specify automatically-determined bins directly in matplotlib, by specifying, e.g. bins='auto'. This uses the maximum of the Sturges and Freedman-Diaconis bin choice.

From [How to choose bin size](https://stackoverflow.com/questions/33458566/how-to-choose-bins-in-matplotlib-histogram)

```python
happyCountry['Happiness.Score'].hist(bins=7, color='lightblue')
```

```python
happyCountry['Happiness.Score'].hist(bins=9, color='blue')
```

```python
happyCountry['Happiness.Score'].hist(bins=12, color='darkblue')
```

```python
happyCountry['Happiness.Score'].hist(bins='auto', color='blue')
```
