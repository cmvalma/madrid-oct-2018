### Import:

```python
import requests, json, pandas as pd

from pandas.io.json import json_normalize
import os
```

### Call:

Request a single question:

```python
rQuest = requests.get('https://opentdb.com/api.php?amount=1')
rQuest
```

Response 200 = success!

Request the list of categories:

```python
rCat = requests.get('https://opentdb.com/api_category.php') 
rCat
```

Request 50 questions (is the m√°ximum allowed) from category "General Knowledge: 

```python
r50GralCat = requests.get('https://opentdb.com/api.php?amount=50&category=9')
r50GralCat
```

```python
r50ScienceCat = requests.get('https://opentdb.com/api.php?amount=50&category=17')
r50ScienceCat
```

# 2. Request data, convert it into a Pandas data frame, and export it as a CSV file. 

### Read and show the Json format:

```python
# data = pd.read_json('https://opentdb.com/api.php?amount=10')
# data
## This would read the Json and put it all inside a col in a Pandas Dataframe. We don't want that
```

```python
questJson = rQuest.json()
display(questJson)
```

```python
catListJson = rCat.json()
display(catListJson)
```

```python
r50GralJson = r50GralCat.json() 
display(r50GralJson)
```

```python
r50ScienceJson = r50ScienceCat.json() 
display(r50ScienceJson)
```

### Normalize the Json:

```python
questNormalized = json_normalize(questJson, 'results')
display(questNormalized)
```

```python
catListNormalized = json_normalize(catListJson, 'trivia_categories')
display(catListNormalized)
```

50 questions of General Knowledge:

```python
r50GralNormalized = json_normalize(r50GralJson, 'results')
display(r50GralNormalized)
```

50 questions of category Science:

```python
r50ScienceNormalized = json_normalize(r50ScienceJson, 'results')
display(r50ScienceNormalized)
```

Both categories joined:

```python
rAllNormalized = r50GralNormalized.append(r50ScienceNormalized)
rAllNormalized
```
